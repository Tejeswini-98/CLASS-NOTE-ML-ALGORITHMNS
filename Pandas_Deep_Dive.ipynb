## Chapter 8 - Grouping Vs Crosstab Vs Pivot-table

```python
insurance_data = pd.read_csv('insurance.csv')
insurance_data
```

```python
insurance_data.groupby(by=['region','smoker'])['charges'].mean()
```

```python
pd.pivot_table(data = insurance_data,values='charges',index='smoker',columns='region',aggfunc='mean',)
```

```python
pd.crosstab(index = insurance_data['smoker'], columns = insurance_data['region'],margins=True) 
#By default computes a frequency table of the factors
```
