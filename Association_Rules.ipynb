```python
saudi_data = online_retail_data[online_retail_data['Country'] == 'Saudi Arabia']
saudi_data
```

```python
saudi_data_pivot_table = pd.pivot_table(data = saudi_data,index='InvoiceNo',columns='Description', values='Quantity').fillna(0)
saudi_data_pivot_table
```

```python
saudi_data_pivot_table.applymap(func = lambda x: 1 if x>=1 else 0)
```

```python
uk_data = online_retail_data[online_retail_data['Country'] == 'United Kingdom']
uk_data
```

```python
uk_data_pivot_table = pd.pivot_table(data = uk_data,index='InvoiceNo',columns='Description', values='Quantity').fillna(0)
uk_data_pivot_table
```

```python
uk_data_encoded = uk_data_pivot_table.applymap(lambda x: 1 if x>=1 else 0)
uk_data_encoded
```

## 5. Model Building

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
freq_item_set = apriori(df = uk_data_encoded,min_support=0.02,use_colnames=True)
freq_item_set
```

```python
best_associates = association_rules(df = freq_item_set ,metric='lift', min_threshold=1)
best_associates
```

```python
best_associates.to_csv('Best_Associates.csv')
```
